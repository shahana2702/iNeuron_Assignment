```python
import pandas as pd
import numpy as np
import string

data={'From_To': ['LoNDon_paris', 'MAdrid_miLAN',
'londON_StockhOlm','Budapest_PaRis', 'Brussels_londOn'],
                   
'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
                   
'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
                   
'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )',
'12. Air France', '"Swiss Air"']}
df = pd.DataFrame(data)
print("_"*35,"Input","_"*35)
print(df.dtypes)
print(df)
print("_"*35,"1","_"*35)
df['FlightNumber'] = df['FlightNumber'].astype('Int64')
for i in df.index:
    if(pd.isna((df.loc[i,'FlightNumber']))):
        df.loc[i,'FlightNumber']=df.loc[i-1,'FlightNumber']+10
print(df)  
print(df.dtypes)
print("_"*35,"2","_"*35)
temp_df=pd.DataFrame(list(df['From_To'].str.split("_")),columns=['A','B'])
print(temp_df)
print("_"*35,"3","_"*35)
for j in temp_df.columns:
    for i in temp_df.index:
        temp_df.loc[i,j]=temp_df.loc[i,j].capitalize()
print(temp_df)
print("_"*35,"4","_"*35)
df[['Country','City']]=temp_df[['A','B']]
df=df.drop('From_To',axis=1)
df=df[['Country','City','FlightNumber','RecentDelays','Airline']]
print(df)

print("_"*35,"5","_"*35)

max_len_item=0
for i in df.index:
    if(max_len_item<len(df.loc[i,'RecentDelays'])):
        max_len_item=len(df.loc[i,'RecentDelays'])   
        
ind=list(string.ascii_uppercase[:max_len_item])

for i in ind:
    df[i]=np.nan

for i in df.index:
    k=0
    for j in range(0,len(df.loc[i,'RecentDelays'])) :
        a=df.loc[i,'RecentDelays']
        df.loc[i,ind[k]]=a[j]
        k=k+1      
print(df)
print("_"*35,"6","_"*35)
df[['RecentDelays','Delay_1','Delay_2']]=df[['A','B','C']]

df=df.drop(['A','B','C'],inplace=False,axis=1)
print(df[['Country','City','FlightNumber','RecentDelays','Delay_1','Delay_2','Airline']])
```
